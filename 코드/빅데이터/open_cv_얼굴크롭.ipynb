```python
# 한국인 안면데이터 얼굴크롭

import glob
import sys
import cv2
import os

# 가중치 파일 경로
cascade_filename = './haarcascade_frontalface_alt.xml'
# 모델 불러오기
cascade = cv2.CascadeClassifier(cascade_filename)
c = 0

for g_f in glob.glob('C:/Users/hope/Downloads/Middle_Resolution/190*'):
    count = 0
    c += 1
    c_n = str(c)
    path = "./Dataset/person_" + c_n
    
    # 사람당 새 폴더 생성
    if not os.path.isdir(path):                                                           
        os.mkdir(path)
    
    for x in range(1,31):
        print(x)
        x_n = str(x)

        for f in glob.glob(g_f+'/S001/L'+x_n+'/E01/C*.jpg'):
            # 이미지 파일
            img = cv2.imread(f)

            # 이미지 압축
            img = cv2.resize(img,dsize=None,fx=1.0,fy=1.0)
            # 그레이 스케일 변환
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
            # cascade 얼굴 탐지 알고리즘 
            results = cascade.detectMultiScale(gray,            # 입력 이미지
                                               scaleFactor= 1.1,# 이미지 피라미드 스케일 factor
                                               minNeighbors=3,  # 인접 객체 최소 거리 픽셀
                                               minSize=(60,60)  # 탐지 객체 최소 크기
                                               )        

            for box in results:
                x, y, w, h = box
                cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255), thickness=2)
                count += 1            
                cv2.imwrite(path+"/User."+c_n+'.'+str(count)+".jpg",img[y:y+h, x:x+w])

                # 사진 출력        
            cv2.imshow('facenet',img)              

if cv2.waitKey(1) == ord('q') : break # q키 입력이 있을 때 반복문 종료
elif count >= 1800 : break #100 face sample
```
